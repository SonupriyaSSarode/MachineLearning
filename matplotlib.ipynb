```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
```

```python

plt.figure(figsize=(15, 15))

#Line Plot
plt.subplot(341)
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title('Line Plot')

#Area Plot
plt.subplot(342)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.fill_between(x, y, alpha=0.5)
plt.title('Area Plot')

#Scatter Plot
plt.subplot(343)
plt.scatter(data[:, 0], data[:, 1])
plt.title('Scatter Plot')

#Bar Plot
plt.subplot(344)
plt.bar(['A', 'B', 'C', 'D'], [10, 20, 15, 25])
plt.title('Bar Plot')

#Histogram
plt.subplot(345)
plt.hist(data[:, 0], bins=20)
plt.title('Histogram')

#Pie Chart
plt.subplot(346)
plt.pie([30, 20, 50], labels=['A', 'B', 'C'])
plt.title('Pie Chart')

#Box Plot
plt.subplot(347)
plt.boxplot(data)
plt.title('Box Plot')

#Heatmap
plt.subplot(349)
plt.imshow(np.corrcoef(data.T), cmap='coolwarm')
plt.colorbar()
plt.title('Heatmap')

#3D Plot
ax = plt.subplot(3,4,10, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2])
plt.title('3D Plot')

#Tis is used to put all plots togther in one plot(Adjust layout and save)
plt.tight_layout()
plt.savefig('all_plots.png')
plt.show()

```
