```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Advertising.csv')
```

```python
print(data)
```

```python
x = data['TV']
y = data['Sales']
```

```python
plt.scatter(x, y)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Linear Regression Using Advertising CSV")
```

```python
def MSE(m, c, points):
    total_error = 0
    for i in range(len(points)):
        x = points.iloc.x
        y = points.iloc.y
        total_error += (y - (m * x + c)) ** 2
    total_error / float(len(points))    
```

```python
def gradient_descent(m_now, c_now, points, L):
    m_gradient = 0
    c_gradient = 0

    n = len(points)

    for i in range(n):
        x = points.iloc[i]['TV']
        y = points.iloc[i]['Sales']

        #calculating gradients
        m_gradient += -(2/n) * x * (y - (m_now * x + c_now))
        c_gradient += -(2/n) *  (y - (m_now * x + c_now))

        #updating parameters after full pass through data
        m = m_now - m_gradient * L
        c = c_now - c_gradient * L
        return m, c
    
#initial parameters    
m = 0
c = 0
L = 0.0001#learning rate
epochs = 1000#number of iterations    

#running gradient descent for epochs number of times
for i in range(epochs):
    m, c = gradient_descent(m, c, data, L)

#output the final parameters
print("slope m =", m)
print("c =", c)

#plotting result
plt.scatter(data['TV'], data['Sales'], color='black')
plt.plot(list(range(int(data['TV'].min()), int(data['TV'].max()))), 
         [m * x + c for x in range(int(data['TV'].min()), int(data['TV'].max()))], 
         color="red")
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Linear Regression Fit with Gradient Descent')
plt.show()
```
