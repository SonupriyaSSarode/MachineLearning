```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix, recall_score,precision_score,f1_score
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.iloc[:, 0:-1], df.iloc[:, -1], test_size = 0.2)
clf1 = LogisticRegression()
```

```python
clf1.fit(xtrain, ytrain)
ypred = clf1.predict(xtest)
print("Accuracy of Logitic Regression = \n", accuracy_score(ytest, ypred))
print("-"*30)
print(pd.DataFrame(confusion_matrix(ytest,ypred),columns=list(range(0,2))))
# print("Confusion matrix is =\n", confusion_matrix(ytest, ypred))
print("-"*30)
print("Precision -\n ",precision_score(ytest,ypred))
print("Recall -\n ",recall_score(ytest,ypred))
print("F1 score - \n",f1_score(ytest,ypred))
```
